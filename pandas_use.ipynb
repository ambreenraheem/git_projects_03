# 16- Splitting a string into  multiple columns

```python
import pandas as pd
df=pd.DataFrame({'name':['Muhammmad Amir','Fahad','Umer','Muhammad Kashif'],
                 'location':['Hyderabad,Pakistan','Karachi,Pakistan','Sanghar,Pakistan','Lahore,Pakistan']})
df
```

```python
# this one is only giving you values. splitting into two columns.
df.name.str.split(' ',expand=True).head()
```

```python
# adding splitting columns into new columns.
df[["first_name","last_name"]] = df.name.str.split(' ',expand=True)
df
```

```python
# not working this time will try later
df[["city","country"]] = df.location.str.split(' ',expand=True)
df
```
